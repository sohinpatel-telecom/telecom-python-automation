```python
sequence_numbers = []

with open("rtp_log.txt", "r") as file:
    for line in file:
        if "Seq=" in line:
            seq = int(line.strip().split("=")[1])
            sequence_numbers.append(seq)

sequence_numbers.sort()

lost_packets = []

for i in range(len(sequence_numbers) - 1):
    current_seq = sequence_numbers[i]
    next_seq = sequence_numbers[i + 1]

    if next_seq != current_seq + 1:
        missing = list(range(current_seq + 1, next_seq))
        lost_packets.extend(missing)

print("Total RTP Packets:", len(sequence_numbers))
print("Lost RTP Packets:", lost_packets)
print("Packet Loss Count:", len(lost_packets))
```
