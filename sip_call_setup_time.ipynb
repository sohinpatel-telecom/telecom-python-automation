```python
invite_time = None
ok_time = None

with open("sip_log.txt", "r") as file:
    for line in file:
        if "INVITE" in line and invite_time is None:
            invite_time = line.split()[0]

        if "200 OK" in line:
            ok_time = line.split()[0]

if invite_time and ok_time:
    invite_sec = float(invite_time.split(":")[-1])
    ok_sec = float(ok_time.split(":")[-1])

    call_setup_time = ok_sec - invite_sec
    print("Call Setup Time:", call_setup_time, "seconds")
else:
    print("INVITE or 200 OK not found in log")
```
