```python
ims_log_content = """12:00:01 SIP: REGISTER sip:ims.com
12:00:01 SIP: 401 Unauthorized
12:00:02 SIP: REGISTER sip:ims.com
12:00:02 SIP: 200 OK
"""

with open("ims_registration_log.txt", "w") as file:
    file.write(ims_log_content)

print("ims_registration_log.txt file created")
```

```python
flow = []

with open("ims_registration_log.txt", "r") as file:
    for line in file:
        if "REGISTER" in line:
            flow.append("REGISTER")
        elif "401" in line:
            flow.append("401")
        elif "200 OK" in line:
            flow.append("200")

print("Detected IMS Flow:", flow)

if flow == ["REGISTER", "401", "REGISTER", "200"]:
    print("IMS Registration Successful (VoLTE Flow)")
elif flow == ["REGISTER", "200"]:
    print("IMS Registration Successful (MCPTT Flow)")
else:
    print("IMS Registration Failed or Incomplete")
```
